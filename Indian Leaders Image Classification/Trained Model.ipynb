```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.applications import imagenet_utils
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline
```

```python
loaded_model = tf.keras.models.load_model("Image_classifier.h5")
```

```python
loaded_model.summary()
```

```python
label = ["APJ Abdul Kalam","Atal Bihari Vajpayee","Chandra Babu Naidu","Dr. B. R. Ambedkar",
          "Indira Gandhi","Lata Mangeshkar","Ratan Tata",
          "Narendra Modi","Vallabhbhai Patel"]
```

```python
label[0]
```

```python
def image_preprocess(image_file):
    img_path = r"D:/Data Science/Deep Learning/image_classification/sample_dataset/"
#    Image(filename = img_path + image_file)
    img = image.load_img(img_path + image_file, target_size = (224, 224))
    img_arr = image.img_to_array(img)
    img_expand_arr = np.expand_dims(img_arr, axis = 0)
    return preprocess_input(img_expand_arr)
```

```python
Image(filename = 'D:\Data Science\Deep Learning\image_classification\sample_dataset/Image_43.JPG')
```

```python
predict = np.argmax(loaded_model.predict(image_preprocess("Image_43.JPG")), axis = 1)
label[predict.item()]
```

```python
predict = np.argmax(loaded_model.predict(image_preprocess("Image_474.JPG")), axis = 1)
print(label[predict.item()])
```
