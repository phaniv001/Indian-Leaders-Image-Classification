```python
#Download images from bing
#1. Ratan Tata
#2. APJ Abdul Kalam
#3. Dr. B. R. Ambedkar
#4. Vallabhbhai Patel
#5. Indira Gandhi
#6. Atal Bihari Vajpayee
#7. Lata Mangeshkar
#8. 
from bing_image_downloader import downloader

#downloader.download("monkey", limit=200,  output_dir='downloads', 
#                    adult_filter_off=True, force_replace=False, timeout=60, verbose=False)
#downloader.download("tiger", limit=200,  output_dir='downloads', 
#                    adult_filter_off=True, force_replace=False, timeout=60)

def Image_download(data):
    for person in data:
        print(person)
        downloader.download(person, limit=2000,  output_dir='sample_dataset', 
                    adult_filter_off=True, force_replace=False, timeout=20, verbose=False)
```

```python
Image_download(["Ratan Tata", "APJ Abdul Kalam", "Dr. B. R. Ambedkar", "Vallabhbhai Patel", "Indira Gandhi", 
                "Atal Bihari Vajpayee", "Lata Mangeshkar", "Chandra Babu Naidu", "Narendra Modi"])
```
